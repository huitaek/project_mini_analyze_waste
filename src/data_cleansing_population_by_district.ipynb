```python
# 행정구역 확인
df['행정구역(시군구)별'].unique()
```

```python
# ~출장소 라는 이름을 가진 행정구역 구분
remove_index = df[df['행정구역(시군구)별'].str.endswith('출장소')].index
print(df.shape[0],remove_index.__len__())
```

```python
# 삭제할 컬럼 확인
print(df['항목'].unique())
print(df['단위'].unique())

remove_columns = ['항목','단위','Unnamed: 138']
```

```python
filled_zero = df.fillna(value=0)
df_removed = filled_zero.loc[:,~filled_zero.columns.isin(remove_columns)].iloc[~df.index.isin(remove_index),:]

print(df_removed.columns.tolist()) # 컬럼 제외 확인
print(df_removed.isna().sum().sum()) # na 값 0으로 대체되었는지 확인
print(df_removed.shape[0]) # 행정구역 이름이 '출장소'로 되어있는 행 제외 확인
```

```python
# 데이텉타입 변경
df_removed.iloc[:,1:] = df_removed.iloc[:,1:].apply(lambda x: x.astype(int),axis=1)
df_removed.to_csv(get_file_dir('refined','행정구역(시군구)별_인구수_tidy.csv'))
df_removed.info()
```

```python
df_removed.head(3)
```

```python
# column 명 정제작업
new_columns = ['행정구역(시군구)',*df_removed.columns[1:].map(lambda x:x.replace(' 월','')).tolist()]

renamed = df_removed.copy()
renamed.columns = new_columns
renamed
```

```python
# restruct
melted = pd.melt(frame=renamed,id_vars=['행정구역(시군구)'],value_vars=renamed.columns[1:]).sort_values(['행정구역(시군구)','variable'])
melted.columns = ['행정구역','연도','명']
melted
```

```python
# 추가 정제작업
# 세종시 => 동일한 데이터, 따라서, 하나는 삭제
index_to_remove = melted[melted['행정구역'].str.startswith('세종시')].index
drop_sejong = melted.drop(index=index_to_remove,axis=0)

drop_sejong[drop_sejong['행정구역'].str.startswith('세종시')]
```

```python
# 대분류(광역지방자치단체) 분류
conds = (
        drop_sejong['행정구역'].str.startswith('전국') 
        | drop_sejong['행정구역'].str.endswith('광역시') 
        | drop_sejong['행정구역'].str.endswith('특별시') 
        | drop_sejong['행정구역'].str.endswith('자치시') 
        | drop_sejong['행정구역'].str.endswith('도')
        )

drop_sejong[conds].to_csv(get_file_dir('refined','인구수_시도별.csv'),index=False)

# 행정구역 확인 
district_gu = drop_sejong[~conds]['행정구역'].unique()
district_gu
```

```python
# 중분류(기초지방자치단체) 분류 및 저장 - 군구
drop_sejong[drop_sejong['행정구역'].isin(district_gu)].to_csv(get_file_dir('refined','인구수_군구별.csv'),index=False)
```
